```python
# df["Reviews"] = df["Reviews"].str.replace("\xa0","").str.strip()

df["Reviews"] = pd.to_numeric(df["Reviews"],errors = "coerce")
median_value = df["Reviews"].median()
df["Reviews"] = df["Reviews"].fillna(median_value)
df["Reviews"] = df["Reviews"].astype(int)
```

```python
df["Rating"] = pd.to_numeric(df["Rating"],errors = "coerce")
median_value  = df["Rating"].median()
df["Rating"] = df["Rating"].fillna(median_value)
df["Rating"] = df["Rating"].astype(int)
```

```python
df["Type"] = pd.to_numeric(df["Type"],errors = "coerce")
median_value = df["Type"].median()
df["Type"] = df["Type"].fillna(median_value)
df["Type"] = df["Type"].astype(str)
```

```python
# Using Imputation method
df["Content Rating"] = df["Content Rating"].astype(str)
imputer = SimpleImputer(strategy = "most_frequent")
df["Content Rating"] = imputer.fit_transform(df[["Content Rating"]]).ravel()
```

```python
df["Current Ver"] = pd.to_numeric(df["Current Ver"],errors = "coerce")
imputer = SimpleImputer(strategy = "most_frequent")
df["Current Ver"] = imputer.fit_transform(df[["Current Ver"]]).ravel()
```

```python
# Step 1: Convert the column to string type to avoid the error
df["Android Ver"] = df["Android Ver"].astype(str)

# Step 2: Now split the string to get the first part (version number)
df["Android Ver"] = df["Android Ver"].str.split(" ").str[0]

# Step 3: Convert to numeric, coercing errors to NaN
df["Android Ver"] = pd.to_numeric(df["Android Ver"], errors='coerce')

# Step 4: Impute missing values (NaNs) using the most frequent value
imputer = SimpleImputer(strategy="most_frequent")
df["Android Ver"] = imputer.fit_transform(df[["Android Ver"]])

# Step 5: Convert to integer after imputation
df["Android Ver"] = df["Android Ver"].astype(float)

# Check result
print(df["Android Ver"].head())
```

# outliers

```python
# function to calculate IQR 
def find_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers =df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# detecting outliers in rating
rating_outliers = find_outliers("Rating")
print(f"Outlier in Rating:\n{rating_outliers[["App","Rating"]]}")

# Detecting outliers in Size (make sure 'Size' is numeric)
df["Size"] = pd.to_numeric(df["Size"],errors = "coerce")

size_outliers = find_outliers("Size")
print(f"Outliers in Size:\n{size_outliers[['App', 'Size']]}")
```

# chi-square test 

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(df["Content Rating"],df["Category"])
# perform chi-square test
chi2_stat,p_value,dof,expected = chi2_contingency(contingency_table)

print(f"chi2-square_statistics :{chi2_stat}")
print(f"p_value: {p_value}")
print(f"dof:{dof}")

if p_value < 0.05:
   print("There is a significant association between Content Rating and Category.")
else:
    print("There is no significant association between Content Rating and Category.")
```
